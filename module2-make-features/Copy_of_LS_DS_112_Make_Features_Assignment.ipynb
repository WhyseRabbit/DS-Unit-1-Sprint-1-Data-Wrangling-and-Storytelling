```python
loan=pd.read_csv('LoanStats_2018Q4.csv',header=1,skipfooter=2,engine='python')
print(loan.shape)
loan.head()
```

```python
private_loan=loan.drop(['url','member_id','id','desc'],axis=1)
```

```python
single_rate=' 16.14%'
float(single_rate.strip('%').strip())
```

```python
type(single_rate)
```

```python
def float_rate(int_rate):
  return float(int_rate.strip('%').strip())
```

```python
private_loan['int_rate']=private_loan['int_rate'].apply(float_rate)
private_loan.head()
```

```python
private_loan['emp_title'].value_counts(dropna=False).head(20)
```

```python
sam_title=['ceo','AcToR','ASTRONAUT','PrinCESS',np.NaN]
```

```python
def clean_title(title):
  if isinstance(title,str):
    return title.strip().title()
  else:
    return 'Unknown'

for example in sam_title:
  print(clean_title(example))
```

```python
private_loan['emp_title']=loan['emp_title'].apply(clean_title)
```

```python
private_loan['emp_title'].value_counts(dropna=False).head(20)
```

```python

private_loan['earliest_cr_line']=pd.to_datetime(private_loan['earliest_cr_line'],
                                           infer_datetime_format=True)

private_loan['earliest_cr_line'].head()
```
